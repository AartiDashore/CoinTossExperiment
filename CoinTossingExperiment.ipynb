```python
import random
import matplotlib.pyplot as plt
```

```python
class CoinTossExperiment:
    def __init__(self, target_heads=3, num_experiments=1000):
        """
        Initialize the experiment with the number of heads needed and the number of experiments to perform.
        Args:
        - target_heads (type: int): Number of heads to observe (here, default is 3).
        - num_experiments (type: int): Total number of experiments to perform (here, default is 1000).
        """
        self.target_heads = target_heads
        self.num_experiments = num_experiments
        self.total_tosses = 0
        self.average_tosses = []

    def toss_until_target_heads(self):
        """
        This method Simulates the coin tosses until the target number of heads is observed.
        Returns:
        - tosses (typeint): The number of tosses needed to get the target number of heads.
        """
        heads_count = 0
        tosses = 0
        while heads_count < self.target_heads:
            tosses += 1
            if random.choice(['H', 'T']) == 'H':  #coin tosses (H for heads, T for tails)
                heads_count += 1
        return tosses

    def perform_experiment(self):
        """
        The coin toss experiment is performed multiple times here and the average number of tosses are counted.
        """
        for i in range(1, self.num_experiments + 1):
            tosses = self.toss_until_target_heads()
            self.total_tosses += tosses
            # Append the average tosses after each experiment
            self.average_tosses.append(self.total_tosses / i)

    def plot_experiment_results(self):
        """
        This method plots the average number of tosses to get the target number of heads over the course of the experiments.
        """
        # Plot every 10th experiment to reduce overhead
        plt.plot(range(10, self.num_experiments + 1, 10), self.average_tosses[9::10], marker='o', linestyle='-', color='blue')
        plt.xlabel('Experiment Number')
        plt.ylabel('Average Number of Tosses')
        plt.title(f'Average Number of Tosses to Get {self.target_heads} Heads')
        plt.show()
```

```python
if __name__ == "__main__":
    # Create an instance of CoinTossExperiment with 3 heads and 1000 experiments
    experiment = CoinTossExperiment(target_heads=3, num_experiments=1000)
    experiment.perform_experiment()
    experiment.plot_experiment_results()
```
